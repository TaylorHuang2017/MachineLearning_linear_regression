## 1.5 计算矩阵乘法 AB

```python
# TODO 计算矩阵乘法 AB，如果无法相乘则raise ValueError

# 两个矩阵相乘，结果应该是i行j列

def matxMultiply(A, B):
    if len(A[0])!= len(B):
        raise ValueError
        return
    p = [[0] * len(B[0]) for i in range(len(A))]
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                p[i][j] += A[i][k] * B[k][j]            
    return p
```

$ A = \begin{bmatrix}

\end{bmatrix} , b = \begin{bmatrix}
    b_{1}  \\

\end{bmatrix}$

返回 $ Ab = \begin{bmatrix}
    a_{11}    & a_{12} & ... & a_{1n} & b_{1}\\
    a_{21}    & a_{22} & ... & a_{2n} & b_{2}\\
    a_{31}    & a_{22} & ... & a_{3n} & b_{3}\\
    ...    & ... & ... & ...& ...\\
    a_{n1}    & a_{n2} & ... & a_{nn} & b_{n} \end{bmatrix}$

```python
# TODO 构造增广矩阵，假设A，b行数相同
# 以下代码可通过zip函数优化 return [ra + rb for ra, rb in zip(A, b)] 

def augmentMatrix(A, b):
    if len(A) != len(b):
        raise ValueError
        return
    
    # result表示最后的矩阵    
    result = []
    for i in range(len(A)):
        # row 是每一行的最终内容
        row = []
        for j in range(len(A[i])):
            row.append(A[i][j])
        for k in range(len(b[i])):
            row.append(b[i][k])
        result.append(row)
    return result
```

## 2.3  Gaussian Jordan 消元法求解 Ax = b

### 2.3.1 算法

步骤1 检查A，b是否行数相同

步骤2 构造增广矩阵Ab

步骤3 逐列转换Ab为化简行阶梯形矩阵 [中文维基链接](https://zh.wikipedia.org/wiki/%E9%98%B6%E6%A2%AF%E5%BD%A2%E7%9F%A9%E9%98%B5#.E5.8C.96.E7.AE.80.E5.90.8E.E7.9A.84-.7Bzh-hans:.E8.A1.8C.3B_zh-hant:.E5.88.97.3B.7D-.E9.98.B6.E6.A2.AF.E5.BD.A2.E7.9F.A9.E9.98.B5)
    
    对于Ab的每一列（最后一列除外）
        当前列为列c
        寻找列c中 对角线以及对角线以下所有元素（行 c~N）的绝对值的最大值
        如果绝对值最大值为0
            那么A为奇异矩阵，返回None (你可以在选做问题2.4中证明为什么这里A一定是奇异矩阵)
        否则
            使用第一个行变换，将绝对值最大值所在行交换到对角线元素所在行（行c） 
            使用第二个行变换，将列c的对角线元素缩放为1
            多次使用第三个行变换，将列c的其他元素消为0
            
步骤4 返回Ab的最后一列

**注：** 我们并没有按照常规方法先把矩阵转化为行阶梯形矩阵，再转换为化简行阶梯形矩阵，而是一步到位。如果你熟悉常规方法的话，可以思考一下两者的等价性。

### 2.3.3 实现 Gaussian Jordan 消元法

```python
# TODO 实现 Gaussain Jordan 方法求解 Ax = b

""" Gaussian Jordan 方法求解 Ax = b.
    参数
        A: 方阵 
        b: 列向量
        decPts: 四舍五入位数，默认为4
        epsilon: 判读是否为0的阈值，默认 1.0e-16
        
    返回列向量 x 使得 Ax = b 
    返回None，如果 A，b 高度不同
    返回None，如果 A 为奇异矩阵
"""
# 以下代码参考了直播通讲--P4,p5 高斯消元和奇异矩阵
# 以下代码参考了[InternetShortcut]
URL=http://localhost:8888/notebooks/Desktop/linear_algebra-master/linear_regression_project.ipynb#
Github project: https://github.com/chilyoniki/Udacity_ML_P3/blob/master/linear_regression_P3/linear_regression_project-2-3.ipynb


def gj_Solve(A, b, decPts=4, epsilon=1.0e-16):
    if len(A) != len(b):  #检查 A， b 行数是否相同
        raise ValueError   # 如果不相同，返回ValueError
        return
    
    Ab = augmentMatrix(A,b)  # 构造增广矩阵 Ab
    
    for c in range(len(Ab[0])-1):  # 逐列进行检查
        Ab_t = transpose(Ab)   # 为了便于寻找列c中对角线以及对角线以下所有元素的绝对值的最大值，先转置矩阵，把列变成行
        c_list = Ab_t[c]     # 取要比较的元素
        max_val = max(c_list[c:],key=abs) # 求绝对值最大的那个数，保存到max_val
        if abs(max_val) < epsilon:    # 如果绝对值是属于0的阈值， 按规则A是奇异矩阵，返回None
            return None
        max_row = c + c_list[c:].index(max_val)  # 记录绝对值最大的数所在的行
        swapRows(Ab,c,max_row)   # 使用第一个行变换，将绝对值最大值所在行交换到对角线元素所在行 （行c）
        scaleRow(Ab,c,1.0/Ab[c][c])   # 使用第二个行变换，将列c的对角线元素缩放为1
        for r in range(len(A)):           # 遍历所有行
            if r != c and Ab[r][c] != 0:    # 除去第 c 行 & 除数不为零
                addScaledRow(Ab,r,c,-Ab[r][c])  # 多次使用第三个变换，将列c的其他元素消为0
    matxRound(Ab)    # 每个元素四舍五入，到小数第四位
    
    return [[val] for val in transpose(Ab)[-1]]  # 返回Ab的最后一列，即方程组的解
```

### 3.2.2 计算平均平方误差 (MSE)

我们要编程计算所选直线的平均平方误差(MSE), 即数据集中每个点到直线的Y方向距离的平方的平均数，表达式如下：
$$
MSE = \frac{1}{n}\sum_{i=1}^{n}{(y_i - mx_i - b)^2}
$$

```python
# TODO 实现以下函数并输出所选直线的MSE
def calculateMSE2D(X,Y,m,b):
    result = 0
    for i in range(0, len(X)):
        result += (Y[i]-m*X[i]-b)**2
        
    return result / len(X)

# TODO 检查这里的结果, 如果你上面猜测的直线准确, 这里的输出会在1.5以内
print(calculateMSE2D(X,Y,m1,b1))
```

$$

### 3.4 求解 $X^TXh = X^TY$ 

在3.3 中，我们知道线性回归问题等价于求解 $X^TXh = X^TY$ (如果你选择不做3.3，就勇敢的相信吧，哈哈)

```python
# TODO 实现线性回归
# XtX is A, Xty is b, 线性回归就是求解 Ax=b
# 先根据 X, Y 求解出 X矩阵，Y矩阵，再求 Xtx 和 Xty，然后利用高斯消元求 Ax=b中的 x
'''
参数：X, Y 存储着一一对应的横坐标与纵坐标的两个一维数组
返回：线性回归的系数(如上面所说的 m, b)
'''
def linearRegression2D(X,Y):
    X = [[v, 1] for v in X]
    Y = [[v] for v in Y]
    X_t = transpose(X)
    A = matxMultiply(X_t, X)
    b = matxMultiply(X_t, Y)
    result = gj_Solve(A, b)    
    return tuple([v[0] for v in result])
```
